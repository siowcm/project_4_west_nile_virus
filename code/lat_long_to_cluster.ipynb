```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
train = pd.read_csv('../input_data/train.csv')
train.head()
```

```python
plt.scatter(train['Longitude'], train['Latitude'], alpha=0.1);
```

```python
kmeans = KMeans()
kmeans.fit(train[['Longitude', 'Latitude']])
train['cluster'] = kmeans.labels_
```

```python
sns.scatterplot(x='Longitude', y='Latitude', data=train, hue='cluster', palette='bright');
```

```python
test = pd.read_csv('../input_data/test.csv')
test.head()
```

```python
test['cluster'] = kmeans.predict(test[['Longitude', 'Latitude']])
```

```python
sns.scatterplot(x='Longitude', y='Latitude', data=test, hue='cluster', palette='bright');
```
